```python
# UDF for calculating vif value
def vif_cal(input_data, dependent_col):
    vif_df = pd.DataFrame( columns = ['Var', 'Vif'])
    x_vars=input_data.drop([dependent_col], axis=1)
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]] 
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=sm.OLS(y,x).fit().rsquared  
        vif=round(1/(1-rsq),2)
        vif_df.loc[i] = [xvar_names[i], vif]
    return vif_df.sort_values(by = 'Vif', axis=0, ascending=False, inplace=False)
```

# Dropping Variable with high VIF

```python
# Calculating Vif value
vif_cal(input_data=zombie_data.drop(['Addresses'],axis=1),dependent_col='Zombie')
```

# Predictions

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logsk = LogisticRegression()
logsk.fit(X_train,y_train)
```

```python
# Predicted probabilities
y_pred = logsk.predict_proba(X_test)
```

```python
# Converting y_pred to a dataframe which is an array
y_pred_df = pd.DataFrame(y_pred)
```

```python
# Converting to column dataframe
y_pred_1 = y_pred_df.iloc[:,[1]]
```

```python
# Let's see the head
y_pred_1.head()
```

```python
y_pred_1.tail()
```

```python
# Converting y_test to dataframe
y_test_df = pd.DataFrame(y_test)
```

```python
# Removing index for both dataframes to append them side by side 
y_pred_1.reset_index(drop=True, inplace=True)
y_test_df.reset_index(drop=True, inplace=True)
```

```python
# Appending y_test_df and y_pred_1
y_pred_final = pd.concat([y_test_df,y_pred_1],axis=1)
```

```python
# Renaming the column 
y_pred_final= y_pred_final.rename(columns={ 1 : 'Zombie_Prob'})
```

```python
y_pred_final
```

```python
# Creating new column 'predicted' with 1 if Zombie_Prob>0.5 else 0
y_pred_final['predicted'] = y_pred_final.Zombie_Prob.map( lambda x: 1 if x > 0.5 else 0)
```

```python
# Let's see the head
y_pred_final.head()
```

# Model Evaluation

```python
from sklearn import metrics
```

```python
help(metrics.confusion_matrix)
```

```python
# Confusion matrix 
confusion = metrics.confusion_matrix( y_pred_final.Zombie, y_pred_final.predicted)
confusion
```

```python
#Let's check the overall accuracy.
metrics.accuracy_score( y_pred_final.Zombie, y_pred_final.predicted)
```
